# Dealing with Missing Data

```python
import pandas as pd
import numpy as np
```

```python
from numpy import NaN
```

```python
data1 = {'Name' :['Person1','person2','person3','person4'], 'Age':[28,27,NaN,36]}
data1
```

```python
data = pd.DataFrame(data1)
data
```

```python
data.isnull()  #check wherther is data Null if its null it will return True
```

```python
data.isnull().sum()   #Return total number of null values
```

```python
data.dropna() #return new dataframe without null value
# data.dropna(inplace=True)   #return original dataframe without null value
```

```python
data['Age'].dtype   # it will always return float becoz it has NuN value
```

```python
data2 = {'Name' :['Person1','person2','person3','person4'], 'Age':[28,27,45,36]}

df=pd.DataFrame(data2)
df
```

```python
df['Age'].dtype  #it will always give int because there is no NuN value
```

# Filling missing values with mean/median/mode

```python
# It fill the missing values of age at index 2

data['Age'].fillna(data['Age'].mean())    #return new dataframe without null value
#data['Age'].fillna(data['Age'].mean(), inplace=True)   #return original dataframe without null value
```

```python
data
```

```python
# No null value in df so no changes of filling null values
df['Age'].fillna(data['Age'].mean())
```

```python
data3 = {'Name' :['Person1','person2','person3','person4'], 'Age':[28,27,27,36]}

df1=pd.DataFrame(data3)
df1
```

```python
df1['Age'].mode()
```

```python
df1['Age'].mode()[0]  #return which has more number of occurence
                      #0 means it will return 1st more occurance value
```

```python
df1['Age'].fillna(df1['Age'].mode()[0])
#df1['Age'].fillna(df['Age'].mode()[0]  inplace=True)
```

```python
df1['Age'].fillna(df1['Age'].median())
```
